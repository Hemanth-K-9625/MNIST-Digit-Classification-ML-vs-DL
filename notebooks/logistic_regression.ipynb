# 🔢 Digit Classification using Logistic Regression
Baseline model for handwritten digit recognition.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

## 📊 Dataset
- MNIST handwritten digits  
- 28×28 grayscale images  
- 10 classes (0–9)

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target)
```

## 🤖 Model (Logistic Regression)
- Linear classifier  
- Suitable for baseline comparison  

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

## 🏋️ Training
- Optimized using gradient descent  
- Trained on flattened image vectors  

```python
model.fit(X_train, y_train)
```

## 📈 Results
- Accuracy: ~95.777777%  
- Fast and simple model  

```python
model.score(X_test, y_test)
```

```python
plt.matshow(digits.images[45])
print("Actual Value :")
print(digits.target[45])
print("Predicted Value :")
model.predict([digits.data[45]])
```

```python
model.predict(digits.data[0:5])
```

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
```

```python
import seaborn as sns
sns.heatmap(cm,annot = True)
plt.xlabel("Predict")
plt.ylabel("Truth")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_predicted,y_test))
```

```python
import joblib
joblib.dump(model, "logistic_model.pkl")
```

## 🏁 Conclusion
Logistic Regression works as a good baseline but cannot capture non-linear relationships.
Simple and fast, but limited performance → motivates use of ANN and CNN.
