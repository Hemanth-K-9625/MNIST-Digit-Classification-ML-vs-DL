# рџ§  Handwritten Digit Classification using Artificial Neural Network (ANN)

## рџ“Њ Objective
The goal of this notebook is to classify handwritten digits (0вЂ“9) using an Artificial Neural Network (ANN) and evaluate its performance on the MNIST dataset.

---

## рџљЂ Why ANN?
- Captures **non-linear relationships**
- More powerful than traditional ML models like Logistic Regression
- Serves as a bridge between classical ML and deep learning

```python
import tensorflow as tf 
from tensorflow import keras
import matplotlib.pyplot as plt 
%matplotlib inline
import numpy as np
```

## рџ“Љ Dataset
- MNIST handwritten digits  
- 28Г—28 grayscale images  
- 10 classes (0вЂ“9)

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
len(X_train)
```

```python
len(X_test)
```

## вљ™пёЏ Preprocessing
- Normalize pixel values (0вЂ“1)  
- Flatten images (784 features)  

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train_flat = X_train.reshape(len(X_train),28*28)
X_test_flat = X_test.reshape(len(X_test),28*28)
X_test_flat.shape
```

```python
plt.matshow(X_train[2])
plt.show()
```

## рџЏ—пёЏ Model (ANN)
- Fully connected layers  
- ReLU activation  
- Sigmoid output (10 classes)

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation = "sigmoid")
])
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train_flat,y_train,epochs = 5)
```

```python
model.evaluate(X_test_flat, y_test)
```

## рџ¤– Training
- Optimizer: Adam  
- Loss: Sparse categorical crossentropy  
- Trained for multiple epochs  

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(100, input_shape=(784,), activation = "relu"),
    keras.layers.Dense(10, activation = "sigmoid")
])
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train,y_train,epochs = 5)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
y_predi = [np.argmax(i) for i in y_pred ]
plt.matshow(X_test[0])
plt.show()
y_predi[0]
```

## рџ“€ Results
- Accuracy: ~98%  
- Better than Logistic Regression  

```python
from sklearn.metrics import classification_report
print(classification_report(y_predi,y_test))
```

```python
model.save("models/ann_model.h5")
```

## рџЏЃ Conclusion
ANN learns non-linear patterns but struggles with spatial structure в†’ CNN performs better.
ANN improves performance, but CNN is the best choice for image tasks.
